Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(25, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(19, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(9, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(9, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(9, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(13, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(13, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.
df_mat_acertos_porcentagem = df_mat['Qt_acertos'].sum()/(24)
print(f"O percentual de acerto em matemática foi de {df_mat_acertos_porcentagem:.2f} %")
````

1. crie uma cópia do *DataFrame* original, contendo somente os registros de vacinas realizadas a menores de 18 anos.  
2. crie uma lista das colunas desse *DataFrame* com o atributo de *DataFrame* **.columns()**

````python
# 1)
df_menores_18 = df[df['paciente_idade'] < 18].copy()
df_menores_18
````

````python
# 4)
colunas_estabelecimento = [
    'estabelecimento_valor', 
    'estabelecimento_razaoSocial', 
    'estabelecimento_municipio_codigo', 
    'estabelecimento_municipio_nome', 
    'estabelecimento_uf', 
    'estalecimento_noFantasia'
]
````

````python
df_estabelecimentos_unicos = df_estabelecimentos_menores_18.drop_duplicates()

df_vacinas_por_estabelecimento = df_menores_18.groupby(colunas_estabelecimento).size().reset_index(name='quantidade_vacinas')

print(df_vacinas_por_estabelecimento)
````
